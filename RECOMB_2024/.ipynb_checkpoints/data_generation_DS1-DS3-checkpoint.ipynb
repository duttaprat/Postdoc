```python
import pandas as pd
import numpy as np
```

```python
def seq2kmer(seq):
    """
    Convert original sequence to kmers
    
    Arguments:
    seq -- str, original sequence.
    k -- int, kmer of length k specified.
    
    Returns:
    kmers -- str, kmers separated by space
    """
    k=6
    kmer = [seq[x:x+k] for x in range(len(seq)+1-k)]
    kmers = " ".join(kmer)
    return kmers
```

```python
df_dev_path = "/data/projects/Enhancer/RECOMB_2024/dev.tsv"
```

```python
df= pd.read_csv(df_dev_path, sep=" ")
df
```

```python
df_kmer= pd.DataFrame()
```

```python
df_kmer['Sequence'] = df['sequence'].apply(seq2kmer)
df_kmer['Label'] =  df['label'].astype(int)
```
