```python
import pandas as pd
```

```python
df = pd.read_csv("/data/private/pdutta/DNABERT_data/Core_Prom_new/TATA/dev.tsv", sep="\t")
```

```python
df
```

```python
# Count the number of 0s and 1s
counts = df['Label'].value_counts(normalize=True)
```

```python
# Calculate the number of 0s and 1s to sample
n_zeros = int(round(counts[0] * 40))
n_ones = 40 - n_zeros  # Ensures the total is 160
```

```python
# Sample the rows
sample_zeros = df[df['Label'] == 0].sample(n=n_zeros)
sample_ones = df[df['Label'] == 1].sample(n=n_ones)
```

```python
# Concatenate the samples to get the final sample DataFrame
final_sample = pd.concat([sample_zeros, sample_ones]).sample(frac=1)  # Shuffle the rows
```

```python
final_sample
```

```python
final_sample.to_csv("/data/private/pdutta/DNABERT_data/Core_Prom_new/TATA/small_dataset/dev.tsv", sep="\t", index=False)
```
