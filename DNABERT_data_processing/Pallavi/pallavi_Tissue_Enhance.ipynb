```python
import os, tarfile
import re
import pandas as pd
import numpy as np
import zipfile, gzip
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
input_path = "/data/projects/Tissue_specific_data/TissueEnhanced/DNABERT_raw/"
```

```python
tissue_name = "brain"
```

```python
df_house = pd.read_csv(input_path+"HouseTrans.csv", sep=",")
df_other = pd.read_csv(input_path+"TspTrans_"+tissue_name+".csv", sep=",")
```

```python
output_folder_path ="/data/projects/Tissue_specific_data/TissueEnhanced/DNABERT_Finetuned_Data/"+tissue_name
if not os.path.exists(output_folder_path + '/Evaluation_data'):
    os.makedirs(output_folder_path + '/Evaluation_data')
```

```python
df_other
```

```python
df_merge = pd.concat([df_house, df_other]).reset_index(drop=True)
df_merge = df_merge.rename(columns={"kmer": "Sequence", "label": "Label"})
df_merge
```

```python
train, test = train_test_split(df_merge, test_size=0.20)
test, val = train_test_split(test, test_size=0.5)
print (train.shape, test.shape, val.shape)
print(train.groupby(['Label']).size(), test.groupby(['Label']).size() , val.groupby(['Label']).size())
```

```python
train[['Sequence','Label']].to_csv(output_folder_path+"/train.tsv", sep="\t", index=False)
test[['Sequence','Label']].to_csv(output_folder_path+"/dev.tsv", sep="\t", index=False)
val[['Sequence','Label']].to_csv(output_folder_path+"/Evaluation_data/dev.tsv", sep="\t", index=False)
```

```python
split_raw_folder = input_path + "/" + tissue_name
if not os.path.exists(split_raw_folder):
    os.makedirs(split_raw_folder)
```

```python
train.to_csv(split_raw_folder+"/train_raw.tsv", sep="\t", index=False)
test.to_csv(split_raw_folder+"/dev_raw.tsv", sep="\t", index=False)
val.to_csv(split_raw_folder+"/test_raw.tsv", sep="\t", index=False)
```
