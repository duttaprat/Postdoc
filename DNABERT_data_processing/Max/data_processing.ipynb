```python
import os
import pandas as pd
import numpy as np
```

```python
non_coding_region= "acceptors"
df= pd.read_csv(f"/data/projects/JimData/DNABERTfiles/splicesites/MOUSE_canonical_{non_coding_region}.tsv", sep="\t", names=[f'{non_coding_region}_sequence'])
```

```python
df
```

```python
def seq2kmer(seq):
    """
    Convert original sequence to kmers
    
    Arguments:
    seq -- str, original sequence.
    k -- int, kmer of length k specified.
    
    Returns:
    kmers -- str, kmers separated by space
    """
    k=6
    kmer = [seq[x:x+k] for x in range(len(seq)+1-k)]
    kmers = " ".join(kmer)
    return kmers
```

```python
df['Sequence']= df[f'{non_coding_region}_sequence'].apply(seq2kmer)
df
```

```python
df['Label'] = np.random.choice([0, 1], size=len(df))
df
```

```python
output_path = f"/data/private/pdutta_new/Max/{non_coding_region}/"
if not os.path.exists(output_path):
    os.makedirs(output_path)
df[['Sequence', 'Label']].to_csv(output_path+ "dev.tsv", sep="\t", index=False)
```
