```python
import pandas as pd
import numpy as np
import pybedtools 
from Bio import SeqIO
from io import StringIO
pybedtools.helpers.set_tempdir("/data/projects/temp")
```

```python
cage_df_38_bed =pd.read_csv("/data/private/pdutta/DNABERT_2/CAGE_seq_hg38/CAGE_seq_hg38_bed.tsv", sep="\t")
cage_df_38_bed
```

```python
pred_non_promoter_df = pd.read_csv("/data/private/pdutta/DNABERT_2/DNABERT2_Prom_Pred/Predicted_Non_Prom_bed.tsv", sep="\t")
pred_non_promoter_df
```

```python
non_promoter_bed = pybedtools.BedTool.from_dataframe(pred_non_promoter_df.iloc[1:])
cage_38_bed = pybedtools.BedTool.from_dataframe(cage_df_38_bed.iloc[1:])
```

```python
intersect_bed = non_promoter_bed.intersect(cage_38_bed, wa=True, wb=True)
```

```python
intersect_df = intersect_bed.to_dataframe()
new_columns = ['chr_name', 'NonCoreProm_start', 'NonCoreProm_end', 'chrom', 'CAGE_start', 'CAGE_end']
intersect_df.columns = new_columns
intersect_df
```

```python
intersect_df.to_csv("/data/private/pdutta/DNABERT_2/intersected_chr22_CAGE_NonCoreProm.tsv", sep="\t", index=False)
```
