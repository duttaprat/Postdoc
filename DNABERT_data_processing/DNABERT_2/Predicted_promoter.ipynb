```python
import numpy as np
import pandas as pd
```

```python
prom_df = pd.read_csv("/data/private/pdutta/DNABERT_2/chr22_all_info.csv", sep=",")
pred_array = np.load('/data/private/pdutta/DNABERT_2/DNABERT2_Prom_Pred/preds.npy')
```

```python
prom_df
```

```python
if len(pred_array) == len(prom_df):
    prom_df['DNABERT_Prediction'] = pred_array
else:
    print("Error: The length of the NumPy array does not match the DataFrame.")
    print(len(prom_df), len(pred_array))
```

```python
prom_df
```

```python
selected_prom = prom_df[prom_df['DNABERT_Prediction']==1][['Coordinates']]
selected_prom
```

```python
selected_prom.insert(0, 'Chromosome', 'chr22')
selected_prom[['Start', 'End']] = selected_prom['Coordinates'].str.split('-', expand=True)
```

```python
selected_prom.drop('Coordinates', axis=1, inplace=True)
selected_prom
```
