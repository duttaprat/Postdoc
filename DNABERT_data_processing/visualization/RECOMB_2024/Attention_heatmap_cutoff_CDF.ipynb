```python
attention_cutoff_path = os.path.dirname(attention_score_path)+"/Visualization/Attention_Cutoff/CDF"
if not os.path.exists(attention_cutoff_path):
    os.makedirs(attention_cutoff_path)
```

```python
def find_subsequent_regions_above_percentile(arr, percentile_val, min_length=3):
    # Calculate the desired percentile for each row
    row_percentiles = np.percentile(arr, percentile_val, axis=1)
    regions = []

    for i, row in enumerate(arr):
        cutoff = row_percentiles[i]
        title = df_visualize['coord'].iloc[i]
        #print(title)
        start, end = map(int, title.split(":")[1].split("-"))
        chr_name = title.split(":")[0]
        # Find indices where values are >= cutoff
        above_cutoff_indices = np.where(row >= cutoff)[0]
        #print(above_cutoff_indices)
        
        # Find subsequent regions of indices
        if len(above_cutoff_indices) > 0:
            start_idx = above_cutoff_indices[0]
            end_idx = start_idx
            
            current_regions = []
            
            for idx in above_cutoff_indices[1:]:
                if idx == end_idx + 1:  # Check if this index is subsequent
                    end_idx = idx
                else:
                    if (end_idx - start_idx + 1) >= min_length:
                        current_regions.append((chr_name , start + start_idx, start+ end_idx, title))
                    start_idx = idx
                    end_idx = idx
            
            if (end_idx - start_idx + 1) >= min_length:
                current_regions.append((chr_name , start+ start_idx, start+ end_idx, title))
            regions.extend(current_regions)
        else:
            print("Check the region ", title)
        #print(regions)
        #input()
    # Convert to DataFrame
    df_result = pd.DataFrame(regions, columns=['chr_name', 'attention_start', 'attention_end', 'coord'])
    
    return df_result
```

```python
for cut_off_value in [70, 75, 80, 85, 90, 95]:
    cut_off_bed_file = find_subsequent_regions_above_percentile(positive_attention_score, cut_off_value)
    cut_off_bed_file.to_csv(attention_cutoff_path+"/attention_BED_cutoff_{}.tsv".format(cut_off_value) , sep="\t", index=False)
```
