```python
folder_path = "/data/projects/Enhancer/Finetuned_data/CPG2_greater_all"
file_extension = '.tsv'
substring = 'encode'
```

```python
def find_files_with_substring_in_name(folder_path, file_extension, substring):
    matching_files = []

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(file_extension) and substring in file:
                matching_files.append(os.path.join(root, file))

    return matching_files
```

```python
result = find_files_with_substring_in_name(folder_path, file_extension, substring)
for file_path in result:
    print(file_path)
    base_name = os.path.basename(file_path)
    name_without_extension = os.path.splitext(base_name)[0]
    new_name = name_without_extension.replace("encode", "encode_new") + os.path.splitext(file_path)[1]
    new_file_path = os.path.join(os.path.dirname(file_path), new_name)
    print(new_file_path)
    df= pd.read_csv(file_path, sep = "\t")
    print(df.shape)
    #print(df['H3K27me3-foreskinfibroblast_0'])
    df_new = df[intersected_feature[0].to_list()]
    #print(df_new)
    df_new.to_csv(new_file_path, sep="\t", index=False)
```
