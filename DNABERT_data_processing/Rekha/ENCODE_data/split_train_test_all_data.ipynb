```python
df = pd.read_csv(train_data_path , sep=",")
#df['Label'] = df['Label'].astype('int')
```

```python
def seq2kmer(seq):
    seq=seq.upper()
    kmer = [seq[x:x+6] for x in range(len(seq)-5)]
    kmers = " ".join(kmer)
    return kmers
```

```python
df_interest = df[['sequence', 'region']]
df_interest
```

```python
df_interest['Label'] = np.where(df_interest['region'] == 'enhancer', 1, 0)
df_interest["Sequence"] = df_interest['sequence'].apply(seq2kmer)
df_interest
```

```python
encode_feature_list  = []

# Open the file in read mode
with open("/data/projects/Enhancer/Data_for_prediction_with_top_features/Encode_top10per_feature.txt", 'r') as file:
    # Loop through the first 300 lines of the file
    for i in range(300):
        line = file.readline()
        # Break the loop if we have reached the end of the file
        if not line:
            break
        # Append the line to the list
        encode_feature_list.append(line.strip())
```

```python
df_encode = df[encode_feature_list]
df_encode
```
