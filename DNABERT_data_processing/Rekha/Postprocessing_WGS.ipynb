```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
output_folder= "/data/projects/Enhancer/Data_results/WholeGenome/chr21/"
```

```python
df= pd.read_csv("/data/projects/Enhancer/Data_results/WholeGenome/chr21/Files_With_PredictedLabels/chr21_GeomicPosition.tsv", sep= "\t")
```

```python
df
```

```python
df_sort = df.sort_values('start').reset_index(drop=True)
```

```python
df_sort
```

```python
df_sort.iloc[-1,3]
```

```python
temp_arr = np.zeros(df_sort.iloc[-1, 3]-df_sort.iloc[0, 2])
```

```python
temp_arr.shape
```

```python
start = df_sort.iloc[0, 2]
```

```python
start
```

```python
for item, row in tqdm(df_sort.iterrows()):
    X= row['start']- start
    Y = row['end']-start # add 1 or not
    temp_arr[X:Y] += float(row['predicted_label']/(Y-X)) 
```

```python
plt.scatterplot()
```

```python
len(temp_arr)
```

```python
with open(output_folder+'prob.npy', 'wb') as f:
    np.save(f, temp_arr)
```
