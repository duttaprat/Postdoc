```python
import pandas as pd
pd.set_option('display.max_columns', None)
import pybedtools
import numpy as np
import os
pybedtools.helpers.set_tempdir('/data/projects/temp')
from IPython.display import display
```

```python
df_exon = df_gtf[df_gtf['feature']=="exon"]
df_exon
```

```python
df_exon.drop_duplicates(['start','end'])
```

```python
df_exon['unique_id'] = df_exon['gene_id'] + '_' + df_exon['start'].astype(str) + '_' + df_exon['end'].astype(str)
```

```python
# Get a list of the existing column names
cols = df_exon.columns.tolist()

# Remove 'start' and 'end' from their current positions in the list
cols.remove('start')
cols.remove('end')

# Insert 'start' and 'end' at the 2nd and 3rd positions
cols.insert(1, 'start')
cols.insert(2, 'end')

# Reorder the DataFrame columns using the modified list of column names
df_exon = df_exon[cols]
```

```python
df_exon
```

```python
df_exon.to_csv("/data/private/pdutta/Collab_data/Jim_work/mouse_gencode_annotation/exon_bed.tsv", sep="\t", index=False)
```

```python
df_intersected_CPD_SNV = pd.read_csv("/data/private/pdutta/Collab_data/Jim_work/Output/UV_Plus/UV_intersected_combined_CPD_SNV.tsv", sep= "\t")
df_intersected_CPD_SNV
```

```python
df_exon_bed = pybedtools.BedTool.from_dataframe(df_exon)
```

```python
df_statistics_intersect = pd.DataFrame(columns=['Distance of SNVs from CPDs', 'unique_CPDs','unique_SNVs', 'unique_EXON', 'unique_Intersected_data'])
df_list = []
for i in range(51):
    df_temp_CPD_SNV = df_intersected_CPD_SNV[abs(df_intersected_CPD_SNV["distance"])==i]
    temp_CPD_SNV_bed = pybedtools.BedTool.from_dataframe(df_temp_CPD_SNV)
    # Perform the intersection
    intersect_CPD_SNV_exon_bed = temp_CPD_SNV_bed.intersect(df_exon_bed, wa=True, wb=True)
    df_intersect_CPD_SNV_exon = intersect_CPD_SNV_exon_bed.to_dataframe()
    df_intersect_CPD_SNV_exon.columns = df_temp_CPD_SNV.columns.to_list() +  ['exon_' + str(x) for x in df_exon.columns.to_list()]
    #display(df_intersect_CPD_SNV_exon)
    print(df_intersect_CPD_SNV_exon.shape)
    df_intersect_CPD_SNV_exon.to_csv("/data/private/pdutta/Collab_data/Jim_work/Output/UV_Plus/Intersected_CPD_SNV_EXON/distance_{}.tsv".format(i), sep="\t", index=False)
    df_list.append(df_intersect_CPD_SNV_exon)
    print("Unique CPDs :", df_intersect_CPD_SNV_exon.drop_duplicates(['CPD_start','CPD_end']).shape)
    print("Unique SNVs :", df_intersect_CPD_SNV_exon.drop_duplicates(['SNV_start','SNV_end','ID',	'SNV_Info']).shape)
    print("Unique exons :", df_intersect_CPD_SNV_exon.drop_duplicates(['exon_start', 'exon_end','exon_gene_id','exon_transcript_id']).shape)
    df_statistics_intersect.loc[i, ['Distance of SNVs from CPDs']] = i
    df_statistics_intersect.loc[i, ['unique_CPDs']] = len(df_intersect_CPD_SNV_exon.drop_duplicates(['CPD_start','CPD_end']))
    df_statistics_intersect.loc[i, ['unique_SNVs']] = len(df_intersect_CPD_SNV_exon.drop_duplicates(['SNV_start','SNV_end','ID',	'SNV_Info']))
    df_statistics_intersect.loc[i, ['unique_EXON']] = len(df_intersect_CPD_SNV_exon.drop_duplicates(['exon_start', 'exon_end','exon_gene_id','exon_transcript_id']))
    df_statistics_intersect.loc[i, ['unique_Intersected_data']] = len(df_intersect_CPD_SNV_exon)
    #input()
```

```python
df_statistics_intersect
```

```python
df_combined = pd.concat(df_list)
df_combined
```

```python
df_combined.to_csv("/data/private/pdutta/Collab_data/Jim_work/Output/UV_Plus/UV_intersected_combined_CPD_SNV_EXON.tsv", sep="\t", index=False)
```

```python
combined_index = len(df_statistics_intersect)+1
df_statistics_intersect.loc[combined_index, ['Distance of SNVs from CPDs']] = "Combined"
df_statistics_intersect.loc[combined_index, ['unique_CPDs']] = len(df_combined.drop_duplicates(['CPD_start','CPD_end']))
df_statistics_intersect.loc[combined_index, ['unique_SNVs']] = len(df_combined.drop_duplicates(['SNV_start','SNV_end','ID',	'SNV_Info']))
df_statistics_intersect.loc[combined_index, ['unique_EXON']] = len(df_combined.drop_duplicates(['exon_start', 'exon_end','exon_gene_id','exon_transcript_id']))
df_statistics_intersect.loc[combined_index, ['unique_Intersected_data']] = len(df_combined)
```
