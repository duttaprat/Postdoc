```python
import os
import numpy as np
import pandas as pd
import sklearn
from io import StringIO
from IPython.display import display
import scipy.stats as stats
```

```python
data_path = "/data/private/pdutta/Collab_data/David_data/GSE91061_fpkm.csv"
```

```python
df  = pd.read_csv(data_path)
df = df.set_index('Unnamed: 0')
df.index.name = None
```

```python
df
```

```python
pre_cols = [col for col in df.columns if 'Pre' in col]
on_cols = [col for col in df.columns if 'On' in col]
```

```python
df_on = df[on_cols]
df_pre = df[pre_cols]
```

```python
#temp_df = df.loc[[]]
temp_df = df_on.loc[[26227, 94081, 6510]]
temp_df
```

```python
temp_df = temp_df.rename(index={26227: 'PHGDH',94081:'SFXN1',6510:'SLC1A5'})
```

```python
df_transposed = temp_df.T
df_transposed['mean'] = df_transposed.mean(axis=1)
df_transposed
```

```python
df_transposed['sample'] = [item.split('_')[0] for item in df_transposed.index.values]
df_transposed.set_index('sample',inplace=True)
```

```python
df_transposed['z_score'] = (df_transposed['mean']-df_transposed['mean'].mean())/df_transposed['mean'].std()
```

```python
df_transposed
```

```python
df_response = pd.read_excel("/data/private/pdutta/Collab_data/David_data/GSE91061_reponse.xlsx")
```

```python
df_response = df_response.set_index('Patient')
df_response.index.name = None
df_response
```

```python
df_merge = pd.merge(df_transposed, df_response, left_index=True, right_index=True)
df_merge
```

```python
response_col = 'On-treatment TCR-Seq'
df_merge = df_merge[['PHGDH', 'SFXN1', 'SLC1A5', 'mean', 'z_score', response_col]]
```

```python
df_merge.groupby(response_col).size()
```

```python
col_name= 'SLC1A5'
```

```python
val_list_3= df_merge[col_name].quantile([.25, .75]).values.tolist()
val_list_3
```

```python
def group_3(row, val_list):
    if row[col_name] < val_list[0]:
        return 0
    elif val_list[0] <= row[col_name] < val_list[1]:
        return 1
    elif val_list[1] <= row[col_name]:
        return 2
```

```python
df_merge['{}_three_clusters'.format(col_name)] = df_merge.apply(lambda row: group_3(row, val_list_3), axis=1)
```

```python
df_merge
```

```python
df_temp = df_merge.groupby([response_col,'{}_three_clusters'.format(col_name)])[col_name].count().reset_index()
df_temp    
```

```python
temp_numpy = df_temp[col_name].to_numpy()
C = np.reshape(temp_numpy, (-1, 3))
C
```

```python
df_temp = df_temp[df_temp['{}_three_clusters'.format(col_name)]!=1]
df_temp
```

```python
temp_numpy = df_temp[col_name].to_numpy()
B = np.reshape(temp_numpy, (-1, 2))
B
```

```python

# importing packages
import scipy.stats as stats
  
#performing fishers exact test on the data
odd_ratio, p_value = stats.fisher_exact(B)
print('odd ratio is : ' + str(odd_ratio))
print('p_value is : ' + str(p_value))
```
