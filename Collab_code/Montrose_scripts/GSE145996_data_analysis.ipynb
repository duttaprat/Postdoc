```python
import os
import numpy as np
import pandas as pd
import sklearn
from io import StringIO
from IPython.display import display
import scipy.stats as stats
```

```python
data_path = "/data/private/pdutta/Collab_data/David_data/GSE145996_FPKM.csv"
```

```python
df  = pd.read_csv(data_path)
df = df.set_index('Gene Name')
df.index.name = None
```

```python
df
```

```python
temp_df = df.loc[['PHGDH','SFXN1','SLC1A5']]
# temp_df = temp_df.iloc[:, 5:]
temp_df.columns = [col.split('_')[-2]+col.split('_')[-1] for col in temp_df.columns]
temp_df
```

```python
df_transposed = temp_df.T
df_transposed['mean'] = df_transposed.mean(axis=1)
df_transposed['z_score'] = (df_transposed['mean']-df_transposed['mean'].mean())/df_transposed['mean'].std()
df_transposed
```

```python
dict_df = pd.read_excel("/data/private/pdutta/Collab_data/David_data/GSE145996_response.xlsx", sheet_name=['Supp1','Supp2'])
df_1  = dict_df.get('Supp1')
df_2  = dict_df.get('Supp2')
df_1 = df_1.set_index('Patient Number')
df_1.index.name = None
df_2 = df_2.set_index('Patient Number')
df_2.index.name = None
df_2 = df_2[df_2['RNASEQ']==1]
df_response = pd.merge(df_1, df_2, left_index=True, right_index=True)
df_response['Best Response' ] = [item[-1] for item in df_response['Best Response']]
df_response['Best Response' ] = [item.replace('D',"NR") for item in df_response['Best Response']]
# df_response['Tumor ID number'] = [item.strip() for item in df_response['sample_ID']]
df_response.set_index('Tumor ID number', inplace=True)
#df_response.index.name = None
```

```python
df_response
```

```python
df_response.groupby('Best Response').size()
```

```python
df_merge = pd.merge(df_transposed, df_response, left_index=True, right_index=True)
df_merge = df_merge.iloc[:, :-1]
df_merge
```

```python
df_merge.groupby('Best Response').size()
```

```python
col_name= 'z_score'
```

```python
val_list_3= df_merge[col_name].quantile([.25, .75]).values.tolist()
val_list_3
```

```python
def group_3(row, val_list):
    if row[col_name] < val_list[0]:
        return 0
    elif val_list[0] <= row[col_name] < val_list[1]:
        return 1
    elif val_list[1] <= row[col_name]:
        return 2
```

```python
df_merge['{}_three_clusters'.format(col_name)] = df_merge.apply(lambda row: group_3(row, val_list_3), axis=1)
```

```python
df_merge
```

```python
df_temp = df_merge.groupby(['Best Response','{}_three_clusters'.format(col_name)])[col_name].count().reset_index()
df_temp    
```

```python
temp_numpy = df_temp[col_name].to_numpy()
C = np.reshape(temp_numpy, (-1, 3))
C
```

```python
df_temp = df_temp[df_temp['{}_three_clusters'.format(col_name)]!=1]
df_temp
```

```python
temp_numpy = df_temp[col_name].to_numpy()
B = np.reshape(temp_numpy, (-1, 2))
B
```

```python

# importing packages
import scipy.stats as stats
  
#performing fishers exact test on the data
odd_ratio, p_value = stats.fisher_exact(B)
print('odd ratio is : ' + str(odd_ratio))
print('p_value is : ' + str(p_value))
```
