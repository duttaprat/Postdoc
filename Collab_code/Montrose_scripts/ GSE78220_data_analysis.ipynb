```python
import os
import numpy as np
import pandas as pd
import sklearn
from io import StringIO
from IPython.display import display
import scipy.stats as stats
```

```python
data_path = "/data/private/pdutta/Collab_data/David_data/GSE78220_PatientFPKM.xlsx"
```

```python
df  = pd.read_excel(data_path)
df = df.set_index('Gene')
df.index.name = None
```

```python
df
```

```python
temp_df = df.loc[['PHGDH','SFXN1','SLC1A5']]
temp_df
```

```python
df_transposed = temp_df.T
df_transposed['mean'] = df_transposed.mean(axis=1)
df_transposed
```

```python
df_transposed['z_score'] = (df_transposed['mean']-df_transposed['mean'].mean())/df_transposed['mean'].std()
```

```python
df_transposed
```

```python
df_response = pd.read_csv("/data/private/pdutta/Collab_data/David_data/GSE78220_response.csv", sep=",")
```

```python
df_response['Patient ID'] = df_response['Patient ID'].astype(str) + '.baseline'
df_response = df_response.set_index('Patient ID')
df_response.index.name = None
```

```python
df_response.groupby('Response').size()
```

```python
df_merge = pd.merge(df_transposed, df_response, left_index=True, right_index=True)
df_merge
```

```python
df_merge.groupby('Response').size()
```

```python
col_name= 'z_score'
```

```python
val_list_3= df_merge[col_name].quantile([.25, .75]).values.tolist()
val_list_3
```

```python
def group_3(row, val_list):
    if row[col_name] < val_list[0]:
        return 0
    elif val_list[0] <= row[col_name] < val_list[1]:
        return 1
    elif val_list[1] <= row[col_name]:
        return 2
```

```python
df_merge['{}_three_clusters'.format(col_name)] = df_merge.apply(lambda row: group_3(row, val_list_3), axis=1)
```

```python
df_merge
```

```python
df_temp = df_merge.groupby(['Response','{}_three_clusters'.format(col_name)])[col_name].count().reset_index()
df_temp    
```

```python
temp_numpy = df_temp[col_name].to_numpy()
C = np.reshape(temp_numpy, (-1, 3))
C
```

```python
df_temp = df_temp[df_temp['{}_three_clusters'.format(col_name)]!=1]
df_temp
```

```python
temp_numpy = df_temp[col_name].to_numpy()
B = np.reshape(temp_numpy, (-1, 2))
B
```

```python

# importing packages
import scipy.stats as stats
  
#performing fishers exact test on the data
odd_ratio, p_value = stats.fisher_exact(B)
print('odd ratio is : ' + str(odd_ratio))
print('p_value is : ' + str(p_value))
```
