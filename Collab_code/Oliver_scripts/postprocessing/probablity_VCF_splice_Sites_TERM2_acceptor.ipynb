```python
import numpy as np
import pandas as pd
import math
```

```python
input_file_path = "/data/private/pdutta/Collab_data/Oliver_data/DNABERT_data/TREM2_Acceptor_Splice_sites/intermediate.tsv"
```

```python
acceptor_probab_file_path  = "/data/private/pdutta/DNABERT_output/Acceptor_Splice_Sites/Prediction_result/TREM2_Acceptor_Splice_sites/pred_results.npy"
```

```python
df_TREM2_Acceptor = pd.read_csv(input_file_path, sep= "\t")
df_TREM2_Acceptor
```

```python
acceptor_probablity = np.load(acceptor_probab_file_path)
```

```python
Reference_acceptor_list = acceptor_probablity[::2]
Alternative_acceptor_list = acceptor_probablity[1::2]
```

```python
df_TREM2_Acceptor['Acceptor_ref']=Reference_acceptor_list
df_TREM2_Acceptor['Acceptor_alt']=Alternative_acceptor_list
```

```python
df_TREM2_Acceptor
```

```python
def calculate_diff_probab(p1,p2):
    return((p2-p1)*max(p1,p2))
```

```python
def log_oods_ratio(p1,p2):
    term1 = math.log2(p1/(1-p1))
    term2 = math.log2(p2/(1-p2))
    return (term1-term2)
```

```python
df_TREM2_Acceptor['AcceptorScoreChange'] = df_TREM2_Acceptor.apply(lambda row: calculate_diff_probab(row['Acceptor_ref'], row['Acceptor_alt']), axis=1)
df_TREM2_Acceptor['AcceptorLogOddRatio'] = df_TREM2_Acceptor.apply(lambda row: log_oods_ratio(row['Acceptor_ref'], row['Acceptor_alt']), axis=1)
```

```python
df_TREM2_Acceptor
```

```python
for i, row in df_TREM2_Acceptor.iterrows():
    s1 = row['Acceptor_sequence']
    s2 = row['Alternative_sequence']
    pos =[i for i in range(len(s1)) if s1[i] != s2[i]]
    if (len(pos)>0):
        print(s1[pos[0]], s2[pos[0]])
        df_TREM2_Acceptor.loc[i, ['Ref_neucleotide']] = s1[pos[0]]
        df_TREM2_Acceptor.loc[i, ['Alt_neucleotide']] = s2[pos[0]]
```
