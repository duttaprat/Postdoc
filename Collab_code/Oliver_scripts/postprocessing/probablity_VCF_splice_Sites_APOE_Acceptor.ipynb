```python
import numpy as np
import pandas as pd
import math
```

```python
input_file_path = "/data/private/pdutta/Collab_data/Oliver_data/DNABERT_data/APOE_Acceptor_Splice_sites/intermediate.tsv"
```

```python
acceptor_probab_file_path  = "/data/private/pdutta/DNABERT_output/Acceptor_Splice_Sites/Prediction_result/APOE_Acceptor_Splice_sites/pred_results.npy"
```

```python
df_APOE_acceptor = pd.read_csv(input_file_path, sep= "\t")
df_APOE_acceptor
```

```python
acceptor_probablity = np.load(acceptor_probab_file_path)
acceptor_probablity
```

```python
Refernce_acceptor_list = acceptor_probablity[::2]
Alternative_acceptor_list = acceptor_probablity[1::2]
```

```python
Refernce_acceptor_list
```

```python
df_APOE_acceptor['Acceptor_ref']=Refernce_acceptor_list
df_APOE_acceptor['Acceptor_alt']=Alternative_acceptor_list
```

```python
df_APOE_acceptor
```

```python
def calculate_diff_probab(p1,p2):
    return((p2-p1)*max(p1,p2))
```

```python
def log_oods_ratio(p1,p2):
    term1 = math.log2(p1/(1-p1))
    term2 = math.log2(p2/(1-p2))
    return (term1-term2)
```

```python
df_APOE_acceptor['AcceptorScoreChange'] = df_APOE_acceptor.apply(lambda row: calculate_diff_probab(row['Acceptor_ref'], row['Acceptor_alt']), axis=1)
df_APOE_acceptor['AcceptorLogOddRatio'] = df_APOE_acceptor.apply(lambda row: log_oods_ratio(row['Acceptor_ref'], row['Acceptor_alt']), axis=1)
```
