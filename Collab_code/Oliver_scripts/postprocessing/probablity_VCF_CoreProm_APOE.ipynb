```python
import numpy as np
import pandas as pd
import math
```

```python
input_file_path = "/data/private/pdutta/Collab_data/Oliver_data/DNABERT_data/APOE_Core_Prom_90BP/all_data.tsv"
```

```python
TATA_probab_file_path  = "/data/private/pdutta/DNABERT_output/TATA_CorePromoter/Prediction_result/APOE_Core_Prom_90BP/pred_results.npy"
nonTATA_probab_file_path  = "/data/private/pdutta/DNABERT_output/nonTATA_CorePromoter/Prediction_result/APOE_Core_Prom_90BP/pred_results.npy"
```

```python
all_data = pd.read_csv(input_file_path, sep= "\t")
all_data
```

```python
df_APOE = all_data.iloc[::2].reset_index(drop=True)
df_APOE
```

```python
df_APOE = df_APOE.drop(['6-mer','TAG'], axis=1)
df_APOE
```

```python
TATA_probablity = np.load(TATA_probab_file_path)
nonTATA_probablity = np.load(nonTATA_probab_file_path)
```

```python
Reference_TATA_list = TATA_probablity[::2]
Alternative_TATA_list = TATA_probablity[1::2]
Refernce_nonTATA_list = nonTATA_probablity[::2]
Alternative_nonTATA_list = nonTATA_probablity[1::2]
```

```python
df_APOE['TATA_ref']=Reference_TATA_list
df_APOE['TATA_alt']=Alternative_TATA_list
df_APOE['nonTATA_ref']=Refernce_nonTATA_list
df_APOE['nonTATA_alt']=Alternative_nonTATA_list
```

```python
df_APOE
```

```python
def calculate_diff_probab(p1,p2):
    return((p2-p1)*max(p1,p2))
```

```python
def log_oods_ratio(p1,p2):
    term1 = math.log2(p1/(1-p1))
    term2 = math.log2(p2/(1-p2))
    return (term1-term2)
```

```python
calculate_diff_probab(0.944215, 1.790375e-03)
```

```python
log_oods_ratio(0.944215, 1.790375e-03)
```

```python
df_APOE['TATAScoreChange'] = df_APOE.apply(lambda row: calculate_diff_probab(row['TATA_ref'], row['TATA_alt']), axis=1)
df_APOE['TATALogOddRatio'] = df_APOE.apply(lambda row: log_oods_ratio(row['TATA_ref'], row['TATA_alt']), axis=1)
df_APOE['nonTATAScoreChange'] = df_APOE.apply(lambda row: calculate_diff_probab(row['nonTATA_ref'], row['nonTATA_alt']), axis=1)
df_APOE['nonTATALogOddRatio'] = df_APOE.apply(lambda row: log_oods_ratio(row['nonTATA_ref'], row['nonTATA_alt']), axis=1)
```

```python
df_APOE
```

```python
df_APOE.to_csv("/data/private/pdutta/Collab_data/Oliver_data/Output/APOE_CoreProm.csv", sep=",",index=False)
```
