```python
import numpy as np
import pandas as pd
import math
```

```python
input_file_path = "/data/private/pdutta/Collab_data/Oliver_data/DNABERT_data/TREM2_Splice_sites/all_data.tsv"
```

```python
acceptor_probab_file_path  = "/data/private/pdutta/DNABERT_output/Acceptor_Splice_Sites/Prediction_result/TREM2_Splice_sites/pred_results.npy"
donor_probab_file_path  = "/data/private/pdutta/DNABERT_output/Donor_Splice_Sites/Prediction_result/TREM2_Splice_sites/pred_results.npy"
```

```python
all_data = pd.read_csv(input_file_path, sep= "\t")
all_data
```

```python
df_TREM2 = all_data.iloc[::2].reset_index(drop=True)
df_TREM2
```

```python
df_TREM2 = df_TREM2.drop(['6-mer','TAG'], axis=1)
df_TREM2
```

```python
acceptor_probablity = np.load(acceptor_probab_file_path)
donor_probablity = np.load(donor_probab_file_path)
```

```python
Reference_acceptor_list = acceptor_probablity[::2]
Alternative_acceptor_list = acceptor_probablity[1::2]
Refernce_donor_list = donor_probablity[::2]
Alternative_donor_list = donor_probablity[1::2]
```

```python
df_TREM2['Acceptor_ref']=Reference_acceptor_list
df_TREM2['Acceptor_alt']=Alternative_acceptor_list
df_TREM2['Donor_ref']=Refernce_donor_list
df_TREM2['Donor_alt']=Alternative_donor_list
```

```python
df_TREM2
```

```python
def calculate_diff_probab(p1,p2):
    return((p2-p1)*max(p1,p2))
```

```python
def log_oods_ratio(p1,p2):
    term1 = math.log2(p1/(1-p1))
    term2 = math.log2(p2/(1-p2))
    return (term1-term2)
```

```python
calculate_diff_probab(0.944215, 1.790375e-03)
```

```python
log_oods_ratio(0.944215, 1.790375e-03)
```

```python
df_TREM2['AcceptorScoreChange'] = df_TREM2.apply(lambda row: calculate_diff_probab(row['Acceptor_ref'], row['Acceptor_alt']), axis=1)
df_TREM2['AcceptorLogOddRatio'] = df_TREM2.apply(lambda row: log_oods_ratio(row['Acceptor_ref'], row['Acceptor_alt']), axis=1)
df_TREM2['DonorScoreChange'] = df_TREM2.apply(lambda row: calculate_diff_probab(row['Donor_ref'], row['Donor_alt']), axis=1)
df_TREM2['DonorLogOddRatio'] = df_TREM2.apply(lambda row: log_oods_ratio(row['Donor_ref'], row['Donor_alt']), axis=1)
```
