```python
import numpy as np
import pandas as pd
import math
```

```python
input_file_path = "/data/private/pdutta/Collab_data/Oliver_data/DNABERT_data/TREM2_Donor_Splice_sites/intermediate.tsv"
```

```python
donor_probab_file_path  = "/data/private/pdutta/DNABERT_output/Donor_Splice_Sites/Prediction_result/TREM2_Donor_Splice_sites/pred_results.npy"
```

```python
df_TREM2_donor = pd.read_csv(input_file_path, sep= "\t")
df_TREM2_donor
```

```python
donor_probablity = np.load(donor_probab_file_path)
donor_probablity
```

```python
Refernce_donor_list = donor_probablity[::2]
Alternative_donor_list = donor_probablity[1::2]
```

```python
Refernce_donor_list
```

```python
df_TREM2_donor['Donor_ref']=Refernce_donor_list
df_TREM2_donor['Donor_alt']=Alternative_donor_list
```

```python
df_TREM2_donor
```

```python
def calculate_diff_probab(p1,p2):
    return((p2-p1)*max(p1,p2))
```

```python
def log_oods_ratio(p1,p2):
    term1 = math.log2(p1/(1-p1))
    term2 = math.log2(p2/(1-p2))
    return (term1-term2)
```

```python
df_TREM2_donor['DonorScoreChange'] = df_TREM2_donor.apply(lambda row: calculate_diff_probab(row['Donor_ref'], row['Donor_alt']), axis=1)
df_TREM2_donor['DonorLogOddRatio'] = df_TREM2_donor.apply(lambda row: log_oods_ratio(row['Donor_ref'], row['Donor_alt']), axis=1)
```

```python
df_TREM2_donor
```

```python
df_TREM2_donor.to_csv("/data/private/pdutta/Collab_data/Oliver_data/Output/TREM2_Donor_Splice_Model.csv", sep=",",index=False)
```
