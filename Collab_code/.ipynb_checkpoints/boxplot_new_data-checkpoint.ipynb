```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
from dash import Dash, dcc, html, Input, Output
```

```python
INPUT_PATH= "/data/projects/shared_data/collab_data/10_transcripts/Data"
INPUT_FOLDERNAME = "tpm"
INPUT_FILENAME = "gbm_gtex.csv"
INPUT_FILENAME1 = "lgg_gtex.csv"
OUTPUT_PATH = "/data/projects/shared_data/collab_data/10_transcripts/Results"
OUTPUT_FILE_NAME = INPUT_FILENAME.split('.')[0]+"_box_plots.png"
```

```python
df = pd.read_csv(INPUT_PATH+"/"+INPUT_FOLDERNAME+"/"+INPUT_FILENAME, sep= " ")
#df = df.T
df
```

```python
df1 = pd.read_csv(INPUT_PATH+"/"+INPUT_FOLDERNAME+"/"+INPUT_FILENAME1, sep= " ")
df1
```

```python
print(df.groupby(['status']).size())
print(df1.groupby(['status']).size())
```

```python
df_merge = pd.concat([df,df1]).drop_duplicates().reset_index(drop=True)
df_merge.rename(columns={'status': 'sample_group'}, inplace=True)
df_merge
```

```python
for column in df_merge.iloc[:,2:].columns:
    print(column)
    min_val = min([float(a) for a in df[column].values.tolist()] )
    fig = px.box(df_merge, x="sample_group", y=column, color= "sample_group")
    #fig.update_traces(quartilemethod="exclusive")
    #fig.update_layout(yaxis_tickformat = '%{y:.2f}')
    #fig.update_layout(yaxis = dict(tickmode = 'linear',tick0 = 5,dtick = 0.25))
    fig.update_yaxes(nticks=10, tickformat=".%2f")
    #fig.update_yaxes(tick0=min_val, dtick=1.50, tickformat='.%2f')
    #fig.update_layout(yaxis_tickformat = '%.2f')
    fig.show()
```

```python
app = Dash(__name__)


app.layout = html.Div([
    html.H4("Analysis of the restaurant's revenue"),
    html.P("x-axis:"),
    dcc.Checklist(
        id='x-axis', 
        options=['status'],
        value=['time'], 
        inline=True
    ),
    html.P("y-axis:"),
    dcc.RadioItems(
        id='y-axis', 
        options=['ENST00000461347.5','ENST00000322723.8','ENST00000484328.1'],
        value='total_bill', 
        inline=True
    ),
    dcc.Graph(id="graph"),
])


@app.callback(
    Output("graph", "figure"), 
    Input("x-axis", "value"), 
    Input("y-axis", "value"))
def generate_chart(x, y):
    df = px.data.tips() # replace with your own data source
    fig = px.box(df, x=x, y=y)
    return fig


app.run_server(debug=True, port=8051)
```
