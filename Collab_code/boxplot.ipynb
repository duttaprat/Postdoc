```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
from dash import Dash, dcc, html, Input, Output
```

```python
INPUT_PATH= "/data/projects/shared_data/collab_data/subsample/boxplot_data"
INPUT_FILENAME = "lgg_gbm_gtex.csv"
OUTPUT_PATH = "/data/projects/shared_data/collab_data/10_transcripts/results"
OUTPUT_FILE_NAME = INPUT_FILENAME.split('.')[0]+"_box_plots.png"
```

```python
df = pd.read_csv(INPUT_PATH+"/"+INPUT_FILENAME, sep= ",", decimal=',', index_col=0, header =0)
df = df.T
df
```

```python
# df = df.astype({'Fee':'float','Discount':'float'})
df = df.astype(float)
df.dtypes
```

```python
df_pheno = pd.read_csv("/data/projects/shared_data/collab_data/subsample/boxplot_data/lgg_gbm_gtex_pheno.csv", sep= ",")
df['sample_group'] = df_pheno['status'].to_list()
```

```python
df
```

```python
for column in df.columns:
    print(column,)
    min_val = min([float(a) for a in df[column].values.tolist()] )
    fig = px.box(df, x="sample_group", y=column, color= "sample_group")
    #fig.update_traces(quartilemethod="exclusive")
    #fig.update_layout(yaxis_tickformat = '%{y:.2f}')
    #fig.update_layout(yaxis = dict(tickmode = 'linear',tick0 = 5,dtick = 0.25))
    #fig.update_yaxes(nticks=10, tickformat=".%2f")
    fig.update_yaxes(tick0=min_val, dtick=1.50, tickformat='.%2f')
    #fig.update_layout(yaxis_tickformat = '%.2f')
    fig.show()
```

```python
app = Dash(__name__)


app.layout = html.Div([
    html.H4("Analysis of the restaurant's revenue"),
    html.P("x-axis:"),
    dcc.Checklist(
        id='x-axis', 
        options=['status'],
        value=['time'], 
        inline=True
    ),
    html.P("y-axis:"),
    dcc.RadioItems(
        id='y-axis', 
        options=['ENST00000461347.5','ENST00000322723.8','ENST00000484328.1'],
        value='total_bill', 
        inline=True
    ),
    dcc.Graph(id="graph"),
])


@app.callback(
    Output("graph", "figure"), 
    Input("x-axis", "value"), 
    Input("y-axis", "value"))
def generate_chart(x, y):
    df = px.data.tips() # replace with your own data source
    fig = px.box(df, x=x, y=y)
    return fig


app.run_server(debug=True, port=8051)
```
