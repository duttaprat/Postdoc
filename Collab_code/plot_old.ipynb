```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

```python
INPUT_PATH= "/data/projects/shared_data/collab_data/10_transcripts"
INPUT_FILENAME = "gbm_gtex.csv"
OUTPUT_PATH = "/data/projects/shared_data/collab_data/10_transcripts/results"
OUTPUT_FILE_NAME = INPUT_FILENAME.split('.')[0]+"_box_plots.png"
```

```python
df = pd.read_csv(INPUT_PATH+"/"+INPUT_FILENAME, sep= ";", decimal=',')
df= df.iloc[: , 1:]
```

```python
df
```

```python
groups_name = list(df.groupby(['status']).groups.keys())
print(groups_name)
```

```python
df_size = df.groupby(['status']).size().to_frame()
df_size
```

```python
x = df.columns.to_list()[2:]
X_lists = []
for (index, rows) in df_size.iterrows():
    print(index,  rows[0])
    temp_list = []
    for item in x:
        temp_list = temp_list + rows[0]* [item]
    print(len(temp_list))
    X_lists.append(temp_list) 
```

```python
len(X_lists[0])
```

```python
Y_lists =[]
for (index, rows) in df_size.iterrows():
    print(index,  rows[0])
    df_temp = df[df['status']==index]
    temp_y_list = []
    for (col_name,col_values) in df_temp.iloc[:, 2:].iteritems():
        temp_y_list.extend(col_values.to_list())
    Y_lists.append(temp_y_list)
```

```python
len(Y_lists[0])
```

```python
fig = go.Figure()
color_list = ['#3D9970' , '#FF4136']
for name, x, y, color in zip(groups_name, X_lists, Y_lists, color_list):
    print(name, len(x), len(y))

    fig.add_trace(go.Box(
        y=y,
        x=x,
        name=name,
        marker_color=color
    ))

fig.update_layout(
    yaxis_title='normalized value',
    boxmode='group' # group together boxes of the different traces for each value of x
)
fig.show()
```

```python
fig.write_image(OUTPUT_PATH+"/plot_"+OUTPUT_FILE_NAME, width=3600, height=600)
```
