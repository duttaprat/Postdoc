```python
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import pickle
```

```python
output_file_path = "/data/private/pdutta/Wang_collaborator/clustreing_result"
```

```python
df = pd.read_excel("/data/private/pdutta/Wang_collaborator/Survival_vs_death_first_day.xlsx")
```

# Clustering the entire data

```python
entire_data = df.iloc[:,3:].to_numpy()
cluster_numbers = [2, 3,4, 5, 6,7,8,9,10,20, 30, 40, 50]
for i in cluster_numbers:
    print(i)
    cluster = KMeans(n_clusters=i).fit(entire_data)
    pickle.dump(cluster, open(output_file_path + '/kmeans_{}.pkl'.format(i), 'wb'))
```

```python
cluster_number = 5
with open(output_file_path+'/kmeans_{}.pkl'.format(cluster_number), 'rb') as f:
    cluster_5 = pickle.load(f)
```

```python
key_list = []
feature_list =np.zeros(cluster_number)
label_list = []
grouped_df =  df.groupby(['MouseNo'])
for key, item in grouped_df:
    #print(key)
    label_list.append(int(grouped_df.get_group(key).iloc[0]['Severity']))
    temp_df = grouped_df.get_group(key).reset_index(drop=True).iloc[:,3:].to_numpy()
    results  = cluster_5.predict(temp_df)
    #print(results)
    u, c = np.unique(results, return_counts=True)
    #print(u, c)
    feature_vec = np.zeros(cluster_number)
    for (i,j) in zip(u, c):
        feature_vec[u]= c
    #print(feature_vec)
    feature_list = np.vstack([feature_list, feature_vec])
    key_list.append(key)
feature_list = np.delete(feature_list, 0, axis=0)
print( feature_list, feature_list.shape)
print(key_list)
print(label_list)
```

```python
feature_list/feature_list.sum(axis=1)[:,None]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(feature_list, np.asarray(label_list), test_size=0.2, random_state=0)
```

```python
print(X_train.shape, Y_train.shape)
```

```python
clf = RandomForestClassifier(n_estimators = 10) 
```

```python
clf.fit(X_train, Y_train)
```

```python
np.argsort(clf.feature_importances_)
```

```python
Y_pred = clf.predict(X_test)
Y_pred
```

```python
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(Y_pred, Y_test))
```

```python
print(metrics.classification_report(Y_test,Y_pred))
```
