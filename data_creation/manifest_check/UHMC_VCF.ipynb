```python
# Initial folder containing BAM files
BAM_FOLDER = "/home/pdutta/Data/Cancer_wiseGDC/Data/Lung/BAM_NEED_TO_BE_PROCESSED/"

# Destination folder where new folders 1, 2, ... will be created
DEST_FOLDER = "/home/pdutta/Data/Cancer_wiseGDC/Data/Lung/NEW_BAM/"

# Create the destination folder if it doesn't exist
os.makedirs(DEST_FOLDER, exist_ok=True)

# Counter for new folder names
count = 1

# Counter for BAM files
bam_count = 1

# Loop over each subfolder in BAM folder
for foldername in os.listdir(BAM_FOLDER):
    folder_path = os.path.join(BAM_FOLDER, foldername)

    # Check if it's a folder
    if os.path.isdir(folder_path):
        
        # Check if the folder contains BAM files
        bam_files = [f for f in os.listdir(folder_path) if f.endswith('.bam')]
        if bam_files:
            # Create a new folder if bam_count is 1
            if bam_count == 1:
                new_folder_path = os.path.join(DEST_FOLDER, str(count))
                os.makedirs(new_folder_path, exist_ok=True)
                count += 1

            # Move BAM folder and its contents to new folder
            shutil.move(folder_path, os.path.join(new_folder_path, foldername))

            # Increment BAM counter
            bam_count += 1

            # Reset BAM counter if it reaches 7 (this will put around 6-7 BAMs in each folder)
            if bam_count > 6:
                bam_count = 1
```
