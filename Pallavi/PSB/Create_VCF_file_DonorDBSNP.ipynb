```python
import pybedtools
```

```python
str_io_donor = StringIO()
df_donor[['chr_name','start','end','transcript_id','strand']].to_csv(str_io_donor, sep='\t', index=False, header=False)
str_io_donor.seek(0)
donor_bed = pybedtools.BedTool(str_io_donor.read(), from_string=True)
```

```python
str_io_transcript = StringIO()
df_transcript[['chromosome_name','transcript_start','transcript_end','ensembl_transcript_id','strand']].to_csv(str_io_transcript, sep='\t', index=False, header=False)
str_io_transcript.seek(0)
transcript_bed = pybedtools.BedTool(str_io_transcript.read(), from_string=True)
```

```python
# intersect the files
intersect_bed = donor_bed.intersect(transcript_bed, wa=True, wb=True)
```

```python
intersect_df = intersect_bed.to_dataframe()
intersect_df.to_csv("/data/private/pdutta/DNABERT_data/Splice_sites/Donor_raw_files/Donor_Transcript_intersect.tsv", sep='\t', index=False)
```
