```python
import os, tarfile
import pandas as pd
import pybedtools
pybedtools.helpers.set_tempdir('/data/projects/temp')
```

```python
gencode_path= "/data/projects/Resources/Gencode_genome_annotation/"
gtf_annotation_filename= "gencode.v40.annotation.gtf"
gencode_gtf_file_path= gencode_path+ gtf_annotation_filename
```

```python
# Define the column names for a GTF file
col_names = ['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute']

# Read the GTF file, skipping lines that start with '#'
df = pd.read_csv(gencode_gtf_file_path, sep='\t', comment='#', names=col_names, header=None)
```

```python
df
```

```python
df = df[['seqname', 'start', 'end', 'source', 'feature','strand', 'frame', 'attribute']]
```

```python
df_gene = df[df['feature']=="gene"]
df_gene
```

```python
df_snp= pd.read_csv("/data/projects/GDC_Cancer_Wise/Brain/Data/Postprocessing_result/Acceptor/important_snp.tsv", sep= "\t")
df_snp
```

```python
split_columns = df_snp.iloc[:, 0].str.split('_', expand=True)

# Create a new DataFrame with the required columns
new_df = pd.DataFrame({
    'chr': split_columns[0],
    'acceptor_start': split_columns[1],
    'acceptor_end': split_columns[2]
})

# Convert 'start' and 'end' columns to numeric type
new_df['acceptor_start'] = pd.to_numeric(new_df['acceptor_start'])
new_df['acceptor_end'] = pd.to_numeric(new_df['acceptor_end'])
```

```python
df_acceptor = new_df.drop_duplicates().reset_index(drop=True)
df_acceptor
```

```python
acceptor_bed = pybedtools.BedTool.from_dataframe(df_acceptor)
gene_bed = pybedtools.BedTool.from_dataframe(df_gene)
```

```python
intersect_gene_acceptor = gene_bed.intersect(acceptor_bed, wa=True, wb=True)
column_list  = df_gene.columns.to_list() + df_acceptor.columns.to_list()
    #print(column_list)
df_intersection = intersect_gene_acceptor.to_dataframe(names=column_list)
```
