```python
import pandas as pd
import os
```

```python
# Load the manifest.tsv file
manifest = pd.read_csv('/data/projects/GDC_Cancer_Wise/Brain/Postprocessing_files/Manifest_Clinical.tsv', sep='\t')

# Base folder containing the subfolders
base_folder = '/data/projects/GDC_Cancer_Wise/Brain/Data/all_vcf_subfolders'
```

```python
manifest
```

```python
# Initialize a list to store file paths
file_paths = []

# Iterate over each row in the manifest DataFrame
for index, row in manifest.iterrows():
    filename = row['filename']
    # Extract the unique identifier from the filename
    unique_id = filename.split('_')[0]
    
    # Initialize a variable to store the file path
    file_path = None
    
    # Search for the file in each subfolder
    for i in range(1, 12):  # Assuming folders are named from 1 to 11
        subfolder = os.path.join(base_folder, f'Folder_{i}')
        for file in os.listdir(subfolder):
            if unique_id in file:
                # Construct the full file path
                file_path = os.path.join(subfolder, file)
                break
        
        # Exit the loop if the file is found
        if file_path:
            break
    
    # Add the file path to the list
    file_paths.append(file_path)

# Add the file_paths list as a new column to the DataFrame
manifest['file_path'] = file_paths


```

```python
manifest[['filename','file_path']]
```

```python
# Save the updated DataFrame to a new TSV file
manifest.to_csv('/data/projects/GDC_Cancer_Wise/Brain/Postprocessing_files/updated_manifest_clinical.tsv', sep='\t', index=False)
```
