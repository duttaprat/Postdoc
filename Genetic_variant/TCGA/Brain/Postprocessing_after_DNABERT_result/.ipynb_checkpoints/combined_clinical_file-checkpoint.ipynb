```python
import pandas as pd
import numpy as np
```

```python
df_gdc_clinical_lgg = pd.read_csv("/data/projects/GDC_Cancer_Wise/Brain/Clinical_files/clinical_LGG.tsv", sep="\t")
df_gdc_clinical_gbm = pd.read_csv("/data/projects/GDC_Cancer_Wise/Brain/Clinical_files/clinical_GBM.tsv", sep="\t")
df_manifest = pd.read_csv('/data/projects/GDC_Cancer_Wise/Brain/Postprocessing_files/Manifest_UUID_CaseUUID.tsv', sep='\t')
```

```python
df_clinical = pd.concat([df_gdc_clinical_gbm, df_gdc_clinical_lgg]).reset_index(drop=True)
df_clinical
```

```python
def aggregate_values(series):
    unique_values = series.dropna().unique()
    if len(unique_values) == 1:
        return unique_values[0]
    else:
        return list(unique_values)
```

```python
# Group by the first three columns and apply the custom aggregation function
df_grouped_clinical = df_clinical.groupby(['case_id', 'case_submitter_id', 'project_id']).agg(aggregate_values).reset_index()
df_grouped_clinical
```

```python
# Merge the dataframes on df_manifest['case_UUID'] and df_grouped_clinical['case_id']
merged_df = pd.merge(df_manifest, df_grouped_clinical, left_on='case_UUID', right_on='case_id', how='left')
merged_df
```

```python
merged_df.groupby(['project_id']).size().reset_index(name='count')
```

```python
# Drop duplicates based on 'case_UUID'
unique_df = merged_df.drop_duplicates(subset='case_UUID')

# Group by 'project_id' in the unique dataframe and count the occurrences
unique_df.groupby('project_id').size().reset_index(name='count')
```

```python
merged_df.groupby(['case_id','case_submitter_id']).agg(aggregate_values).reset_index().to_csv("unique_patient.tsv", sep="\t")
```
