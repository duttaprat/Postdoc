## Only for TCGA Samples

```r
# Initialize the columns with NA values
df$case_UUID <- NA
df$TCGA_barcode <- NA
df$biospecimen <- NA

# Loop through each UUID and update the case_UUID and TCGA_barcode columns
for (i in 1:nrow(df)) {
  tryCatch({
    # Retrieve the case UUID
    case_uuid <- UUIDtoUUID(df$id[i], to_type = "case_id")
    df$case_UUID[i] <- ifelse(length(case_uuid) > 0, case_uuid[2], NA)
    
    # Convert UUID to TCGA barcode
    barcode_info <- UUIDtoBarcode(df$id[i], from_type = "file_id")
    df$TCGA_barcode[i] <- barcode_info$associated_entities.entity_submitter_id[1]
    
    # Use the TCGA barcode to get biospecimen information
    biospec_info <- TCGAbiospec(df$TCGA_barcode[i])
    df$biospecimen[i] <- biospec_info$sample_definition[1]
    
  }, error = function(e) {
    # Leave the columns as NA in case of an error
    df$case_UUID[i] <- NA
    df$TCGA_barcode[i] <- NA
    df$biospecimen[i] <- NA
  })
}
```

## All GDC samples

```r
# Initialize empty lists to store the clinical data for each part
demographic_list <- list()
diagnoses_list <- list()
exposures_list <- list()
main_list <- list()

for (i in 1:nrow(df)) {
  tryCatch({
    # Retrieve the case UUID
    case_uuids <- UUIDtoUUID(df$id[i], to_type = "case_id")
    df$case_UUID[i] <- ifelse(length(case_uuids) > 0, case_uuids[2], NA)
    
    # Convert UUID to TCGA barcode
    barcode_info <- UUIDtoBarcode(df$id[i], from_type = "file_id")
    df$Barcode[i] <- list(barcode_info$associated_entities.entity_submitter_id)
    
      # Check if there are at least two UUIDs and use the second one
    if (!is.null(case_uuids) && length(case_uuids) >= 1) {
      # Extract the UUID as a character string
      case_uuid <- as.character(case_uuids[1, "cases.case_id"])
      print(case_uuid)

      # Proceed only if case_uuid is a non-empty character string
      if (is.character(case_uuid) && nchar(case_uuid) > 0) {
        clindat <- gdc_clinical(case_uuid)
      
        # Append each part of the clinical data to the respective list
        demographic_list[[i]] <- clindat[["demographic"]]
        diagnoses_list[[i]] <- clindat[["diagnoses"]]
        exposures_list[[i]] <- clindat[["exposures"]]
        main_list[[i]] <- clindat[["main"]]
      }
    }
    
  }, error = function(e) {
    # Error handling
    df$case_UUID[i] <- NA
    df$Barcode[i] <- NA
  })
}

# Concatenate the data from each list into separate dataframes
demographic_df <- bind_rows(demographic_list)
diagnoses_df <- bind_rows(diagnoses_list)
exposures_df <- bind_rows(exposures_list)
main_df <- bind_rows(main_list)
```
