```python
# Initial folder containing VCF files
VCF_FOLDER = "/data/projects/GDC_Cancer_Wise/Brain/Data/VCF_multi/"

# Destination folder where new folders 1, 2, ... will be created
DEST_FOLDER = "/data/projects/GDC_Cancer_Wise/Brain/Data/SUB_VCF/"

# Create the destination folder if it doesn't exist
os.makedirs(DEST_FOLDER, exist_ok=True)

# Counter for new folder names
folder_count = 1

# Counter for VCF files
vcf_count = 1

# List to store destination folder paths
dest_folders = [os.path.join(DEST_FOLDER, f"Folder_{i}") for i in range(1, 11)]

# Create 10 subfolders in the destination folder
for folder in dest_folders:
    os.makedirs(folder, exist_ok=True)

# Get a list of all VCF files
vcf_files = [os.path.join(VCF_FOLDER, f) for f in os.listdir(VCF_FOLDER) if f.endswith('.vcf.gz')]

# Loop over each VCF file and move them to destination folders
for vcf_file in vcf_files:
    # Get the destination folder
    dest_folder = dest_folders[(vcf_count - 1) % 10]
    
    # Move the VCF file to the destination folder
    shutil.move(vcf_file, os.path.join(dest_folder, os.path.basename(vcf_file)))
    
    # Increment the VCF file counter
    vcf_count += 1

print(f"Distribution of VCF files completed. Total files processed: {vcf_count - 1}")
```
