```python
import pandas as pd
import pybedtools
print(pybedtools.__file__)
pybedtools.helpers.set_tempdir('/home/pdutta/temp')
```

```python
non_coding_region="acceptor"
```

```python
df= pd.read_csv("/home/pdutta/Data/Noncoding_region/Unique_{}_RAW_Combined_BED.tsv".format(non_coding_region), sep="\t")
df
```

```python
acceptor_bed = pybedtools.BedTool.from_dataframe(df)
```

```python
acceptor_bed_dup = pybedtools.BedTool.from_dataframe(df)
```

```python
intersect_vcf_acceptor = acceptor_bed.intersect(acceptor_bed_dup, wa=True, wb=True)
```

```python
df_intersection = intersect_vcf_acceptor.to_dataframe()
df_intersection
```

```python
# Convert columns to strings if they are not already (to ensure proper comparison)
# Ensuring all necessary columns are in string format for accurate comparison
df_intersection['chr1'] = df_intersection['chr1'].astype(str)
df_intersection['12572'] = df_intersection['12572'].astype(str)
df_intersection['12651'] = df_intersection['12651'].astype(str)
df_intersection['chr1.1'] = df_intersection['chr1.1'].astype(str)
df_intersection['12572.1'] = df_intersection['12572.1'].astype(str)
df_intersection['12651.1'] = df_intersection['12651.1'].astype(str)

# Filtering out rows where values in specified columns are the same
# Filtering out rows where the values in the specified columns are identical
filtered_df = df_intersection[
    (df_intersection['chr1'] != df_intersection['chr1.1']) | 
    (df_intersection['12572'] != df_intersection['12572.1']) | 
    (df_intersection['12651'] != df_intersection['12651.1'])
]
```
