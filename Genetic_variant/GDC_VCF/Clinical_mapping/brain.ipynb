```python
merged_df = pd.merge(mapping_tsv, clinical_tsv, left_on='patient', right_on='case_submitter_id', how='left')
merged_df
```

```python
# Define a function to handle aggregation of columns:
def aggregate_rows(x):
    if x.dtype == 'float64':  # Assuming numerical columns are float64; adjust if necessary
        return x.mean()
    else:
        return x.iloc[0]  
```

```python
aggregated_df = merged_df.groupby(['manifest_patient_id', 'submitter_id']).agg(aggregate_rows).reset_index()
```
