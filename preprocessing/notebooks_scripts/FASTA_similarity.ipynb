```python
import os
import pandas as pd
import sys 
import subprocess
```

```python
cmd = ["blastn", "-query",'../1.txt',"-subject", '../3.txt', "-task", 'blastn', "-outfmt","6"]
object1 = subprocess.Popen(cmd, stdout=subprocess.PIPE)
object1.wait()
```

```python
if sys.version_info[0] < 3: 
    from StringIO import StringIO
else:
    from io import StringIO

temp = StringIO(object1.communicate()[0].decode('utf-8'))
header = ['query', 'ref', 'per_ident', 'aln_len', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']
df = pd.read_csv(temp, sep="\t", names=header)
```
