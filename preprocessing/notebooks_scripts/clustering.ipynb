```python
import sklearn
import pandas as pd
from sklearn.cluster import KMeans
```

```python
data_path = "/data/projects/Tissue_specific_data/clustering/2.5.csv"
```

```python
df = pd.read_csv(data_path)
```

```python
df = df.set_index("tissue")
df
```

```python
clustering_data = df
```

```python
clustering_data
```

```python
cluster = KMeans(n_clusters=3).fit(clustering_data)
```

```python
for i in range(2,6):
    cluster = KMeans(n_clusters=i).fit(clustering_data)
    labels = cluster.labels_
    sil_score = sklearn.metrics.silhouette_score(clustering_data, labels)
    #ch_score = sklearn.metrics.calinski_harabasz_score(clustering_data, labels)
    print("Cluster_{}: ".format(i))
    print("Silhouette Score: ", sil_score)
```
