```python
df_transcript = df_transcript.loc[((df_transcript['transcript_type']=='miRNA') | (df_transcript['transcript_type']=='protein_coding') | (df_transcript['gene_type']=='lncRNA')) & (df_transcript['seqname']!='chrM')]
df_transcript
```

```python
unique_transcripts_list = []
for file in chr_files:
    if file.endswith('.fa'):
        print ("File name:", file)
        chr_name = file.split('.')[0]
        df_temp = df_transcript[df_transcript["seqname"]==chr_name]
        print(df_temp.shape)
        df_temp_sort = df_temp.sort_values(by = 'start')
        #.head(40)
        temp_sort_diff =df_temp_sort[df_temp_sort['start'].diff()>1100]
        unique_transcripts_list.append(temp_sort_diff['transcript_id'].tolist())
```

```python
len(unique_transcripts_list[1])
```

```python
flat_list= [item for sublist in unique_transcripts_list for item in sublist]
len(flat_list)
```

```python
check_df = df1[df1['transcript_id'].isin(flat_list)].reset_index()
check_df
```

```python
check_df.to_csv(data_save_path+"1100bp_promoter_gtf_annotation.csv", index=False, ignore_index=True)
```
